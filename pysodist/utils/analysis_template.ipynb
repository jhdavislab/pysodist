## Setup
* Here we will set all of the relevant paths pointing to your isodist output

```python
wd = '/cold2/jhdavis/pysodist_rewrite/Ace_15N_600_DIA/' #this is your working directory for the sample you're analyzing
ft = '/cold2/jhdavis/pysodist_rewrite/Ace_15N_600_DIA/model_15Npulse_3peaks_isodist_fits/' #this is where your isodist fits are stored
ip = '/cold2/jhdavis/pysodist_rewrite/Ace_15N_600_DIA/model_15Npulse_3peaks_final_plots/interactive_plots/' #where you would like to store any interactive plots
isodist_result = '/cold2/jhdavis/pysodist_rewrite/Ace_15N_600_DIA/model_15Npulse_3peaks_isodist_outputs/model_15Npulse_3peaks_output.csv' # your isodist result file
NUM = ['AMP_U'] #set these fields to the ratio you'd like to inspect.
DEN = ['AMP_U', 'AMP_F']
id_result = ps.parse_isodist_csv(isodist_result)

#####################################################################
id_result = ps.set_current_ratio(id_result, numerator = NUM, denominator = DEN) #always run set_current_ratio before plotting!
ratio_string = '['+'+'.join([num for num in NUM])+']/['+'+'.join([den for den in DEN])+']' #always run update ratio_string before plotting!
######################################################################
```
